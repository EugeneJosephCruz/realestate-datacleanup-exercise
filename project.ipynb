#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO
#normalize the variable of prices using z-score standardization
#"Fuenlabrada"
mean_price_f = ds_south_belt_fuenlabrada['price'].mean()
std_price_f = ds_south_belt_fuenlabrada['price'].std()
ds_south_belt_fuenlabrada['price_zscore'] = (ds_south_belt_fuenlabrada['price'] - mean_price_f) / std_price_f
#"Leganés"
mean_price_l = ds_south_belt_leganes['price'].mean()
std_price_l = ds_south_belt_leganes['price'].std()
ds_south_belt_leganes['price_zscore'] = (ds_south_belt_leganes['price'] - mean_price_f) / std_price_f
#"Getafe" 
mean_price_g = ds_south_belt_getafe['price'].mean()
std_price_g = ds_south_belt_getafe['price'].std()
ds_south_belt_getafe['price_zscore'] = (ds_south_belt_getafe['price'] - mean_price_g) / std_price_g
# "Alcorcón"
mean_price_a = ds_south_belt_alorcon['price'].mean()
std_price_a = ds_south_belt_alorcon['price'].std()
ds_south_belt_alorcon['price_zscore'] = (ds_south_belt_alorcon['price'] - mean_price_a) / std_price_a

dataframes = [ds_south_belt_fuenlabrada, ds_south_belt_leganes, ds_south_belt_getafe, ds_south_belt_alorcon]
colors = ['blue', 'green', 'red', 'purple']
labels = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]
plt.figure(figsize=(8, 6))

for i, df in enumerate(dataframes):
    plt.hist(df['price_zscore'], bins=20, alpha=0.5, color=colors[i], edgecolor='black', label=labels[i])

plt.xlabel('Z-Score Standardized Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price Z-Scores from Multiple DataFrames')

plt.legend()

plt.show()








```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# TODO

populations = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for i, population in enumerate(populations):
    row, col = i // 2, i % 2
    ax = axes[row, col]
    
    population_data = ds_south_belt[ds_south_belt['level5'] == population]
    
    ax.scatter(range(len(population_data)), population_data['price'])
    
    ax.set_title(f'Scatter plot for {population}')
    
    ax.set_xlabel('Index')
    ax.set_ylabel('Price')

plt.tight_layout()

plt.show()

#Combine all into the same graph
populations = {
    'Fuenlabrada': {'marker': 'o', 'color': 'blue', 'label': 'Fuenlabrada'},
    'Leganés': {'marker': 's', 'color': 'green', 'label': 'Leganés'},
    'Getafe': {'marker': 'D', 'color': 'red', 'label': 'Getafe'},
    'Alcorcón': {'marker': '^', 'color': 'purple', 'label': 'Alcorcón'},
}

plt.figure(figsize=(10, 6))

for population, params in populations.items():
    population_data = ds_south_belt[ds_south_belt['level5'] == population]
    plt.scatter(range(len(population_data)), population_data['price'], marker=params['marker'], color=params['color'], label=params['label'])

plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Scatter Plot of Prices for Different Populations')
plt.legend()

plt.show()

```

```python
from ipyleaflet import Map, basemaps

# Map centred on (60 degrees latitude et -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.Stamen.Terrain)
map
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:

# Replace ',' with '.'' in 'latitude' and 'longitude' columns
ds_south_belt['latitude'] = ds_south_belt.loc[:,'latitude'].str.replace(',', '.')
ds_south_belt['longitude'] = ds_south_belt.loc[:,'longitude'].str.replace(',', '.')

# Convert 'latitude' and 'longitude' columns to float
ds_south_belt['latitude'] = ds_south_belt.loc[:,'latitude'].astype(float)
ds_south_belt['longitude'] = ds_south_belt.loc[:,'longitude'].astype(float)

# Print 'latitude' and 'longitude' columns
print(ds_south_belt.loc[:,'latitude':'longitude'])

from ipyleaflet import Marker

#remember: populations = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']


# Extract 'latitude' and 'longitude' columns into a dictionary
location_dict = ds_south_belt[['latitude', 'longitude']].to_dict('records')

# Add markers for each data point from the dictionary
for loc in location_dict:
    marker = Marker(location=(loc['latitude'], loc['longitude']))
    map.add_layer(marker)

# Display the map
map
```
